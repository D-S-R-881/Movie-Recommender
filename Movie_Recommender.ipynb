```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# hame genres mai se for eg. above hame action,adventure, fantasy, science fiction extract karna h toh vo ham ek function bana kar karsakte h lekin
# genres ek string type object h toh pehle hamko usko list banana hoga fir uske upar function apply ho payega
# toh string ko list banane ke liye ham ast library ka use karenge
```

```python
import ast
```

```python
# literal_eval function in ast is used to convert string to list
def convert(text):
  l = []
  for i in ast.literal_eval(text):
    l.append(i['name'])
  return l
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].keywords
```

```python
# From keywords also we will extract names
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
# Now for cast we will extract names of first top 5 cast
def convert_cast(text):
  L = []
  counter = 0
  for i in ast.literal_eval(text):
    if counter < 5:
      L.append(i['name'])
      counter = counter + 1
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(text):
  l = []
  for i in ast.literal_eval(text):
    if i['job'] == 'Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# Now changing the overview column from string to list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# Ab hame name mai se space remove karna hoga because same name ke kaafi bande ho sakte h toh model fir sahi recommendation nahi de payega
# Like for eg. , Johnny Depp -> JohnnyDepp
# Same cheez hame genre mai bhi karni hogi for like, science fiction -> sciencefiction etc.
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['genres'] + movies['overview'] + movies['cast'] + movies['crew'] + movies['keywords']
```

```python
movies.head()
```

```python
df = movies[['movie_id','title','tags']]
```

```python
df.head()
```

```python
# Converting tags columns for list to string
df['tags'] = df['tags'].apply(lambda x : " ".join(x))
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
# Now applying stemming
# ['love','loving','loved']
# ['love','love','love']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y )
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
df.head()
```

```python
# Basically hamko 2 films ke tags ke bich mai simalrity nikaalna h toh ham inko vectors mai change karke similarity nikaalte h basically
# jab hamko similar movies nikaalna hoti h tab ham inn vectors ke sabse pass wali movies recommend karte h. Ab for sabse pass wali movies we
# can use distance between them but distance ek ideal parameter nahi hota h inn cases mai toh ham cosine similarity use karte h
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 7000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# Now calculating cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[1]
```

```python
df[df['title']=='Avatar'].index[0]
```

```python
# Ab ham similarity ko sort karke top ke 8 similarity ko nikaal lenge aur vo hi recommend kar denge
# Lekin sort karne mai similarity ka index lost ho jayega
# Isliye ek enumerate function ko call karke ham (index, value) pair mai list bana denge fir sorting karenge toh index lost nahi hoga
```

```python
sorted(list(enumerate(similarity[0])), reverse=True,key=lambda x:x[1])[1:9]

# lambda x:x[1] means ham 2nd value ke base pe sort kar rahe h
```

```python
def recommend_movie(movie):
  movie_index = df[df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:9]

  for i in movies_list:
    print(df.iloc[i[0]].title)
```

```python
recommend_movie('Avatar')
```

```python
recommend_movie("Harry Potter and the Order of the Phoenix")
```

```python
recommend_movie("Iron Man")
```

```python
recommend_movie('Captain America: Civil War')
```
